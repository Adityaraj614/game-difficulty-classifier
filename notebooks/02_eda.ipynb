```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"]=(8,5)
```

```python
df = pd.read_csv("../data/raw/game_difficulty_data.csv")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["difficulty"].value_counts()
```

```python
sns.countplot(x="difficulty", data=df)
plt.title("Difficulty Class Distribution")
plt.show()
```

```python
sns.boxplot(x="difficulty", y="enemy_density", data=df)
plt.title("Enemy Density vs Difficulty")
plt.show()
```

```python
sns.boxplot(x="difficulty", y="health_pickups", data=df)
plt.title("Health Pickups vs Difficulty")
plt.show()
```

```python
sns.boxplot(x="difficulty", y="accuracy_required", data=df)
plt.title("Accuracy Required vs Difficulty")
plt.show()
```

```python
sns.boxplot(x="difficulty", y="time_limit", data=df)
plt.title("Time Limit vs Difficulty")
plt.show()
```

```python
sns.boxplot(x="difficulty", y="map_complexity", data=df)
plt.title("Map Complexity vs Difficulty")
plt.show()
```

```python
sns.pairplot(
    df,
    hue="difficulty",
    diag_kind="kde",
    plot_kws={"alpha": 0.6}
)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(
    df.drop(columns=["difficulty"]).corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
## EDA Insights

##- Enemy density and time limit show clear
#   separation between Easy and Hard levels.
##- Medium difficulty overlaps with Hard in accuracy_required,
#   indicating ambiguous boundaries.
##- Health pickups inversely correlate with difficulty,
#   validating label logic.
##- Map complexity alone is a weak separator but adds value 
#   when combined with other features.
```
